```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn import preprocessing
from sklearn import pipeline

number = 107

ext = ".txt"

fn = str(number) + ext

#==================INPUT INFORNAMTION===================
#read *.txt file
data = pd.read_csv(str(fn),header = None, names=["Elevation"])

data = pd.DataFrame(data)

print(data)


plt.plot(data)
#plt.imshow(data)
```

```python
#depth/samples
samples = 616
c=3*(10**8)
permit=9.0
t_window = 139.303478 * 10**(-9)
#depth = t_window*c
depth = t_window*(c/np.sqrt(permit))

d_depth = depth / samples
print("depth/samples=", d_depth)
print("+++++++++++++++++++++++++++++++++")
print("")


curve = np.zeros(shape=len(data),dtype=int)

#print(np.info(curve))

print(np.min(data[:]["Elevation"]))
print(data.iloc[1]["Elevation"])
      
for i in range(0,len(data),1):   
    curve[i] = np.round( (data.iloc[i]["Elevation"]  - np.min(data[:]["Elevation"]))/d_depth ) 

plt.plot(curve)
```

```python
fmt2 = "%i"

o_fn = str(number) + "_GPR" + ext

print(o_fn)

np.savetxt(o_fn,curve,fmt=fmt2)
```
