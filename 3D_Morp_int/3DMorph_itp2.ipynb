```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import random

def goto(linenum):
    global line
    line = linenum

ch = pd.read_csv("Ch_ele.txt", delim_whitespace=True,
          header=None, names=["Elev","Index"])
tr = pd.read_csv("trace.txt", delim_whitespace=True,
          header=None, names=["GPR_points"])


N = 110
max = 0

size = np.zeros(shape=(N), dtype=int)
size2 = np.zeros(shape=(N), dtype=int)

for i in np.arange(1,N+1,1):
    tm = ch.loc[ch["Index"]==str(i),["Elev"]]
    size[i-1] = np.size(tm)
    
max = np.max(size)
    
print("max number of GNSS points",max)

for i in np.arange(1,N+1,1):
    size2[i-1] = tr["GPR_points"][i-1]
#    print(size2[i-1])
min = np.min(size2)
print("min number of GPR points",min)

#The shape of final output
Elev_ch4 = np.zeros(shape=(N,min))
Elev_ch5 = np.zeros(shape=(N,min))
```

```python
GNSS = np.zeros(shape=(N,max))

for n in np.arange(1,N+1,1):
#for n in np.arange(51,52,1):
#for n in (51,52):

    print("++++++")
    print("n=",n)
    print("++++++")
    tm = ch.loc[ch["Index"]==str(n),["Elev"]]
    tm_size = np.size(tm)
    print("\"tm_size\" means the number of GNSS points")
    print("tm_size=",tm_size)
#    print(tm)
    for j in np.arange(0,tm_size,1):
        GNSS[n-1][j] = tm["Elev"][j+sum(size[:(n-1)])]
    
###############################################################
    G_point = tr["GPR_points"][n-1]
    print("G_point=",G_point)

    Elev_ch2 = np.zeros(shape=(G_point))
    Elev_ch3 = np.zeros(shape=(G_point))

    tm2 = np.zeros(shape=G_point)

    print("Elev_ch2=",np.size(Elev_ch2))    

    print("G_point/tm_size",G_point/tm_size)
    print("itv=math.ceil(G_point/tm_size) ")
    itv = math.ceil(G_point/tm_size)   #   
    print("itv=",itv)
    
    tm_size2 = math.floor(G_point/itv)
    print("tm_size2=math.floor(G_point/itv)")
    print("tm_size2=",tm_size2)
    
    M = tm_size2 * itv 
    print("M=",M)
    for j in np.arange(0,tm_size2+1,1):       
#        print("j=",j,"j+itv=",j*itv, "GNSS=",GNSS[n-1][j]) #Do not remove this sentence!!
        if(j*itv >= G_point):
            break
        Elev_ch2[j*itv] = GNSS[n-1][j]
        Elev_ch3[j*itv] = GNSS[n-1][j]        

#    if(j*itv >= G_point):
        #break
#        goto(56)
#    else:
#    Elev_ch2[M] = GNSS[n-1][tm_size2+1]       

###############################################################    
    for j in np.arange(itv,(tm_size2+1)*itv,itv):    
#    for j in np.arange(itv,(tm_size2*itv),itv):    

#        print(j)

        if (j == G_point and GNSS[n-1][tm_size2+1] != 0.0 ):
            Elev_ch2[j-1] = GNSS[n-1][tm_size2+1] 
            print("j == G_point","j=",j,"Elev_ch2[",j-1,"]=",Elev_ch2[j-1])

            for i in np.arange(1,itv,1):
                j2 = j- 1
                
                Elev_ch2[(j2-itv)+i] = Elev_ch2[j2-itv] + ( (i*0.05) * (Elev_ch2[j2] - Elev_ch2[j2-itv])/(0.05 * itv) )
        elif(j < G_point):
  
            for i in np.arange(1,itv,1):
                Elev_ch2[(j-itv)+i] = Elev_ch2[j-itv] + ( (i*0.05) * (Elev_ch2[j] - Elev_ch2[j-itv])/(0.05 * itv) )

        elif(j == G_point and GNSS[n-1][tm_size2+1] == 0.0 ):          
            Elev_ch2[j-1] = random.uniform(GNSS[n-1][tm_size2-2],GNSS[n-1][tm_size2-1])
            print(GNSS[n-1][tm_size2-2],GNSS[n-1][tm_size2-1])
            print(Elev_ch2[j-1])
            for i in np.arange(1,itv,1):
                j2 = j- 1
#                print("Elev_ch2[",j2,"], Elev_ch2[",j2-itv,"]=",Elev_ch2[j2], Elev_ch2[j2-itv])
                Elev_ch2[(j2-itv)+i] = Elev_ch2[j2-itv] + ( (i*0.05) * (Elev_ch2[j2] - Elev_ch2[j2-itv])/(0.05 * itv) )
#                print("j2-itv+i=",j2-itv+i,Elev_ch2[(j2-itv)+i])

###############################################################    
    if(M == G_point):
        goto(90)
    else:    
        print("else")
        Elev_ch2[G_point-1] = GNSS[n-1][tm_size-1]
 #       Elev_ch2[G_point-1] = GNSS[n-1][tm_size2+2]
        itv2 = (G_point-1)- M
        print("itv2 = (G_point-1)- M =",itv2)
        print("M = tm_size2 * itv =", M)
        print("Elev_ch2[",M,"]",Elev_ch2[M])
        print("Elev_ch2[",G_point-1,"]",Elev_ch2[G_point-1])
        
        for i in np.arange(1,itv2+1,1):          
            Elev_ch2[M+i] = Elev_ch2[M] + ( (i * 0.05) * (Elev_ch2[G_point-1] - Elev_ch2[M])/(0.05 * itv2) )       
                
###############################################################    
#flip the track at each odd track
#    if((n-1) % 2 != 0) :
#        print (Elev_ch2[3])
        
#        print (Elev_ch3[-4])

    for j in np.arange(0,min,1):
        Elev_ch4[n-1][j] = Elev_ch2[j]
        if((n-1) % 2 != 0):
            Elev_ch3[:] = np.flipud(Elev_ch2[:])           
            Elev_ch4[n-1][j] = Elev_ch3[j]
            
         
#            print(n-1,"flip")
        
#    for j in np.arange(0,min,1):
#        Elev_ch5[n-1][j] = Elev_ch2[j]    
                 
###############################################################    
    plt.plot(Elev_ch2,"o")
#    plt.plot(Elev_ch2[1479:1513],"o")    
#    plt.ylim(1602,1608)
    plt.legend(["51","52"])
```
